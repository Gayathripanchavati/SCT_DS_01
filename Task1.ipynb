## Load the data

### Subtask:
Load the data from "/content/world_population.csv" into a dataframe.


```python
import pandas as pd

df = pd.read_csv('/content/world_population.csv')
display(df.head())
```

## Visualize the data

### Subtask:
Create dashboards using different plots to visualize the data.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style('whitegrid')

# 1. Histogram of '2022 Population'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='2022 Population', bins=50, kde=True)
plt.title('Distribution of 2022 Population')
plt.xlabel('2022 Population')
plt.ylabel('Frequency')
plt.show()

# 2. Scatter plot of 'Area (km²)' versus 'Density (per km²)'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Area (km²)', y='Density (per km²)')
plt.title('Area vs. Density')
plt.xlabel('Area (km²)')
plt.ylabel('Density (per km²)')
plt.show()

# 3. Bar plot of average 'Growth Rate' for each 'Continent'
plt.figure(figsize=(12, 7))
average_growth_rate = df.groupby('Continent')['Growth Rate'].mean().reset_index()
sns.barplot(data=average_growth_rate, x='Continent', y='Growth Rate')
plt.title('Average Growth Rate by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Growth Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Box plot of 'World Population Percentage' by 'Continent'
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='Continent', y='World Population Percentage')
plt.title('World Population Percentage by Continent')
plt.xlabel('Continent')
plt.ylabel('World Population Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   A histogram of the '2022 Population' was generated to show the distribution of population sizes across countries.
*   A scatter plot of 'Area (km²)' versus 'Density (per km²)' was created to visualize the relationship between country area and population density.
*   A bar plot showing the average 'Growth Rate' for each 'Continent' was produced to compare growth trends across continents.
*   A box plot of 'World Population Percentage' by 'Continent' was generated to show the distribution and spread of each continent's contribution to the total world population.
